```python
import tensorflow as tf
import pyspark
from pyspark.sql import SparkSession

print("TESTE PRA VER SE TA TUDO OK")
print(f"Versão do TF: {tf.__version__}")

# Verifica se a GPU tá visível pro python
gpu_devices = tf.config.list_physical_devices('GPU')
if gpu_devices:
    print(f"Status da GPU: ✅ CONECTADA ({len(gpu_devices)} encontrada(s))")
    for gpu in gpu_devices:
        print(f" - Detalhes: {gpu}")
else:
    print("Status da GPU: ❌ NÃO ENCONTRADA")

print("\nIniciando o Motor do Apache Spark...")
# Tenta iniciar uma sessão do Spark (isso testa se o Java foi instalado corretamente)
spark = SparkSession.builder.appName("TesteInicial").getOrCreate()
print(f"Versão do PySpark: {spark.version}")
print("Status do Spark: ✅ RODANDO")
```
